```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import association_rules, apriori
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv('teleco_market_basket.csv')
```

```python
# Example of a single transaction in the dataset
df.iloc[3]
```

```python
df.head()
```

```python
df = df.dropna(how = 'all')
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
# Inspiration came from (2)
lists = []
for i in range(len(df)):
    row = []
    for j in range(len(df.columns)):
        value = str(df.iat[i, j])
        row.append(value)
    lists.append(row)
```

```python
encode = TransactionEncoder()
array = encode.fit(lists).transform(lists)
```

```python
clean = pd.DataFrame(array, columns = encode.columns_)
```

```python
clean
```

```python
for col in clean.columns:
    print(col)
```

```python
clean = clean.drop(['nan'], axis = 1)
```
